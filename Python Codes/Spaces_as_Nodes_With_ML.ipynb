```python
#Utility Function to Convert Locations to Nodes and Vice Versa

def locations_to_nodes(arr):
  start_x = arr[0]
  start_y = arr[1]

  new_start_x = start_x//20
  new_start_y = start_y//20

  return (min(new_start_x, 5), min(new_start_y, 3))

def nodes_to_locations(node1, node2):
  return (node1*20 + 10, node2*20 + 10)
```

```python
#Making the Field with the Network

def generate_passes_network_for_starting11(matchid, team, color, color1):
  graph = nx.DiGraph()
  match_data = pd.DataFrame(sb.events(match_id = matchid, split = True, flatten_attrs = True, fmt = "dict")).transpose()
  passes = pd.DataFrame(match_data["pass"].dropna())
  team_name = match_data.iloc[team]["team"]["name"]
  opponent_name = match_data.iloc[int(not team)]["team"]["name"]
  #print(team_name)
  
  team_index = []
  for index in match_data["possession_team"].index:
    if match_data["possession_team"].loc[index]["name"] == team_name:
      team_index.append(index)
  #print(team_index)

  team_data = match_data.loc[team_index]
  #print(team_data)

  for x_corr in range(6):
    for y_corr in range(4):
      graph.add_node((x_corr,y_corr))

  players = []
  for player in match_data.iloc[team]["tactics"]["lineup"]:
    players.append(player["player"]["name"])

    #graph.add_node(player["player"]["name"])

  weights = defaultdict(lambda: defaultdict(int))
  pass_data = pd.DataFrame(team_data["pass"].dropna())
  pass_index = pass_data.index
  #print(pass_data)
  
  positions = defaultdict(list)
  for index in pass_index:
    try:
      recieve_pass = match_data["player"].loc[index]["name"]
      give_pass = passes["pass"].loc[index]["recipient"]["name"]
      #print(recieve_pass,give_pass)
      if (recieve_pass in players) and (give_pass in players):
        positions[recieve_pass].append(passes["pass"].loc[index]["end_location"])
        positions[give_pass].append(match_data.loc[index]["location"])

        give_pass_loc, recieve_pass_loc = locations_to_nodes(match_data.loc[index]["location"]), locations_to_nodes(passes["pass"].loc[index]["end_location"])
        weights[give_pass_loc][recieve_pass_loc]+=1
        graph.add_edge(give_pass_loc, recieve_pass_loc, weight = 1/(weights[give_pass_loc][recieve_pass_loc]))

      else:
        pass
    
    except KeyError:
      continue

  #avg_positions=defaultdict(int)
  #for p in positions:
  #  avg_positions[p]=np.sum(np.array(positions[p]),axis=0)/len(np.array(positions[p]))
  #  avg_positions[p]=[130*avg_positions[p][0]/120,90-(90*avg_positions[p][1]/80)]


  fig = plt.figure(figsize = (13, 11))
  ax = fig.add_subplot(1, 1, 1)


  plt.plot([0, 0],[0, 90], color = "black")
  plt.plot([0, 130],[90, 90], color = "black")
  plt.plot([130, 130],[90, 0], color = "black")
  plt.plot([130, 0],[0, 0], color = "black")
  plt.plot([65, 65],[0, 90], color = "black")

#Left Penalty Area
  plt.plot([16.5, 16.5], [65, 25], color = "black")
  plt.plot([0, 16.5], [65, 65], color = "black")
  plt.plot([16.5, 0], [25, 25], color = "black")

#Right Penalty Area
  plt.plot([130, 113.5], [65, 65], color = "black")
  plt.plot([113.5, 113.5], [65, 25], color = "black")
  plt.plot([113.5, 130], [25, 25], color = "black")

#Left 6-yard Box
  plt.plot([0, 5.5], [54, 54], color = "black")
  plt.plot([5.5, 5.5], [54, 36], color = "black")
  plt.plot([5.5, 0.5], [36, 36], color = "black")

#Right 6-yard Box
  plt.plot([130, 124.5], [54, 54], color = "black")
  plt.plot([124.5, 124.5], [54, 36], color = "black")
  plt.plot([124.5, 130], [36, 36], color = "black")

#Prepare Circles
  centreCircle = plt.Circle((65, 45), 9.15, color = "black", fill = False)
  centreSpot = plt.Circle((65, 45), 0.8, color = "black")
  leftPenSpot = plt.Circle((11, 45), 0.8, color = "black")
  rightPenSpot = plt.Circle((119, 45), 0.8, color = "black")

#Draw Circles
  ax.add_patch(centreCircle)
  ax.add_patch(centreSpot)
  ax.add_patch(leftPenSpot)
  ax.add_patch(rightPenSpot)

#Prepare Arcs
  leftArc = Arc((11, 45), height = 18.3, width = 18.3, angle = 0, theta1 = 310, theta2 = 50, color = "black")
  rightArc = Arc((119, 45), height = 18.3, width = 18.3, angle = 0, theta1 = 130, theta2 = 230, color = "black")

#Draw Arcs
  ax.add_patch(leftArc)
  ax.add_patch(rightArc)

#Tidy Axes
  plt.axis('off')


  positions=defaultdict(list)
  for node1 in range(6):
    for node2 in range(4):
      positions[(node1, node2)] = list(nodes_to_locations(node1, node2))

  #print(positions)

  
  edges = graph.edges()
  weights = [graph[u][v]['weight'] for u, v in edges]
  plt.title("Passing Network for " + team_name + " in the game between " + opponent_name + " and " + team_name)
  nx.draw_networkx(graph,node_color = color, font_color = color1, width = (np.array(weights)**-1)/2, pos = positions,edge_color = "lightgrey")

  #print(positions)
  print(weights)
  return graph #,avg_positions
```

```python
def classification_report_with_accuracy_score(y_true, y_pred):
    originalclass.extend(y_true)
    predictedclass.extend(y_pred)
    return accuracy_score(y_true, y_pred) # return accuracy score
```

### ***Gathering the Data for ML Models***

```python
#Getting the Data
gd = []

in_m_home = []
in_m_away = []

for [i,j] in np.array(sb.competitions()[["competition_id","season_id"]]):
  try:
    x = sb.matches(competition_id = i, season_id = j)
  except:
    continue
  for k in range(len(x)):
    if (x.iloc[k]['home_score'] - x.iloc[k]['away_score'] != 0):
      #print(i,j,k)
      f = x.iloc[k]['home_score']
      a = x.iloc[k]['away_score']
      m = x.iloc[k]['match_id']
      if f-a < 0:
        gd.append(-1)
      elif f-a > 0:
        gd.append(1)
    
      #Home Team
      g_home = generate_passes_network_data_for_starting11(m, 0)
      measures_home = []
      measures_home.append(nx.average_clustering(g_home))
      measures_home.append(nx.transitivity(g_home))
      #measures.append(nx.average_shortest_path_length(g))
      measures_home.append(sum(nx.degree_centrality(g_home).values()) / len(nx.degree_centrality(g_home)))
      measures_home.append(sum(nx.pagerank(g_home).values()) / len(nx.pagerank(g_home)))
      measures_home.append(sum(nx.betweenness_centrality(g_home).values()) / len(nx.betweenness_centrality(g_home)))
      measures_home.append(sum(nx.closeness_centrality(g_home).values()) / len(nx.closeness_centrality(g_home)))
      in_m_home.append(measures_home)

      #Away Team
      g_away = generate_passes_network_data_for_starting11(m, 1)
      measures_away = []
      measures_away.append(nx.average_clustering(g_away))
      measures_away.append(nx.transitivity(g_away))
      #measures.append(nx.average_shortest_path_length(g))
      measures_away.append(sum(nx.degree_centrality(g_away).values()) / len(nx.degree_centrality(g_away)))
      measures_away.append(sum(nx.pagerank(g_away).values()) / len(nx.pagerank(g_away)))
      measures_away.append(sum(nx.betweenness_centrality(g_away).values()) / len(nx.betweenness_centrality(g_away)))
      measures_away.append(sum(nx.closeness_centrality(g_away).values()) / len(nx.closeness_centrality(g_away)))
      in_m_away.append(measures_away)

    else:
      continue

#Creating the Input and Output Arrays
le = LabelEncoder()
gd = le.fit_transform(gd)

X_home = np.array(in_m_home)
X_away = np.array(in_m_away)
y = np.array(gd)
```

```python
def avg_degree_distribution(team):
  d=sb.matches(competition_id=55, season_id=43)
  match_id=[]
  match_team=[]
  for index in np.array(d[d["home_team"]==team]["match_id"]):
    match_id.append(index)
    match_team.append(0)
  for index in np.array(d[d["away_team"]==team]["match_id"]):
    match_id.append(index)
    match_team.append(1)

  degree_distribution=[]
  for i in range(len(match_id)):
    g=generate_passes_network_data_for_starting11(match_id[i],match_team[i])
    degree_distribution.append(nx.degree_histogram(g))

  avg_degree_dist=np.zeros(300)
  for i in degree_distribution:
    for j in range(len(i)):
      avg_degree_dist[j]+=i[j]
  avg_degree_dist=avg_degree_dist/len(degree_distribution)

  return(avg_degree_dist)
```
